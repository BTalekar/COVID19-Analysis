```python
#import the required libraries
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('uk_data.csv')
data
```

```python
#Data for the first 2 Weeks
data.drop(data.iloc[:,43:], axis = 0, inplace = True)
data
```

```python
data.info()
```

```python
columns = ['2019','2020','2021','2022','2023']
col_sum = data[columns].sum()
print(col_sum)
```

```python
# Plot a horizontal bar graph
for index, value in enumerate(col_sum):
    plt.text(value, index, str(value), ha='right', va='center')
    
plt.barh(columns,col_sum)
plt.xlabel('Deaths')
plt.ylabel('Year')
plt.title('Total Deaths innthe UK under 43 weeks (England and Wales)')
plt.show()
```

### Conclusion : The deaths in England and Wales in 2020 and 2023 are similar
